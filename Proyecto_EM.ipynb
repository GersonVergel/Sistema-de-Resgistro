Cargar Base de datos

```r
read.csv("/content/H.csv")
Datos=read.csv("/content/H.csv")
```

```r
Mujeres=filter(Datos, Sexo=="Mujer")
Mujeres
```

```r
Seleccion=select(Mujeres,Año,Mes_año,Población_alcaldía,Porcentaje_insegura,Percepción.de.Inseguridad_TP)
Seleccion
```

```r
Correlacion= cor(Seleccion, method=("pearson"))
Correlacion

round(Correlacion,4)
```

```r
Correlacion_2= cor.mtest(Seleccion,conf.level=0.95)

Correlacion_2

corrplot(Correlacion)
```

```r
corrplot(Correlacion,method= "square")
```

```r
corrplot(Correlacion,method= "color")
```

```r
corrplot(Correlacion,tl.col = "purple",method= "square", diag = FALSE)
```

```r
corrplot(Correlacion,tl.col = "purple",method= "square", diag = FALSE, tl.cex = 0.9)
```

```r
corrplot(Correlacion,tl.col = "purple",method= "square", diag = FALSE, tl.cex = 0.9, col=brewer.pal(n=10,name="RdYlBu"))
```

```r

corrplot(Correlacion,tl.col = "purple", type="upper",method= "square",
         diag = FALSE, tl.cex = 0.8, col=brewer.pal(n=10,name="RdYlBu"))

corrplot(Correlacion,tl.col = "purple",add=TRUE, type="lower",method= "number",
         diag = FALSE, tl.cex = 0.1, col=brewer.pal(n=10,name="RdYlBu"))
```

```r
read.csv("/content/Z.csv")
Datos2=read.csv("/content/Z.csv")
```

```r
Mujeres2=filter(Datos2, Sexo=="Mujer")
Mujeres2
```

```r

Seleccion2=select(Mujeres2,Año,'Mes.año',Porcentaje,Proporción,Observación,
                  'Población.de.la.alcaldía','Población.que.enfrentó.violencia')
Seleccion2
```

```r
Correlacion2= cor(Seleccion2, method=("pearson"))

round(Correlacion2,4)

Correlacion2_2= cor.mtest(Seleccion2,conf.level=0.95)

Correlacion2_2

corrplot(Correlacion2)
```

```r
corrplot(Correlacion2,method= "square")
```

```r
corrplot(Correlacion2,method= "color")
```

```r
corrplot(Correlacion2,tl.col = "purple",method= "square",
         diag = FALSE, tl.cex = 0.9, col=brewer.pal(n=10,name="RdYlBu"))
```

```r
corrplot(Correlacion2,tl.col = "purple",method= "square",
         diag = FALSE, tl.cex = 0.8,tl.pos="l",col=brewer.pal(n=10,name="RdYlBu"))

corrplot(Correlacion2,tl.col = "purple",add=TRUE, type="lower",method= "number",
         diag = FALSE, tl.cex = 0.8,tl.pos="l",col=brewer.pal(n=10,name="RdYlBu"))
         
```

```r
summary(Seleccion2)

Componentes_Principales= princomp(Seleccion2, cor= TRUE)

Componentes_Principales

summary(Componentes_Principales)

```

```r
#calcular la varianza total explicada por cada componente principal
var_explained = Componentes_Principales $ sdev ^ 2 / sum (Componentes_Principales $ sdev ^ 2)

qplot (c (1: 7), var_explained,col="red") + 
  geom_line () + 
  xlab (" Componente principal ") +
  ylab (" Explicación de la varianza ") +
  ggtitle (" Diagrama de Scree ") +
  ylim (0, 1)
```

```r
summary(var_explained)
```

```r
#Se toman 2 componentes principales

unclass(loadings(Componentes_Principales))

plot(Componentes_Principales, scale=0, col= "green")
```

```r
Modelo = hclust(dist(Mujeres2), method="ward.D2")

plot(Modelo)

rect.hclust(Modelo, k=3, border="magenta")
```

```r
Modelo = hclust(dist(Seleccion2), method="ward.D2")

plot(Modelo)

rect.hclust(Modelo, k=3, border="magenta")
```

```r
Dendograma= as.dendrogram(Modelo)

labels_colors(Dendograma) <- 1:5

plot(Dendograma, main = "Dendograma")
```

```r
Dendograma <- Dendograma %>%
  color_branches(k = 7) %>%
  set("branches_lwd", c(2,1,2)) %>%
  set("branches_lty", c(1,2,1))

plot(Dendograma)
```
